```python
import cv2
import matplotlib.pyplot as plt
import pytesseract
import numpy as np
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'
```

```python
img=cv2.imread('video/image1.jpg')
```

```python
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
score=img[:25,1555:1620]
plt.imshow(score)
cv2.imwrite('score.jpg',score)
str_score = pytesseract.image_to_string(score)
print(str_score)
```

```python
kda=img[:25,1665:1710]
plt.imshow(kda)
cv2.imwrite('kda.jpg',kda)
str_kda = pytesseract.image_to_string(kda, config='outputbase digits')
print(str_kda)
```

```python
time=img[:25,1860:1905]
plt.imshow(time)
cv2.imwrite('time.jpg',time)
str_time = pytesseract.image_to_string(time, config='outputbase digits')
print(str_time)
```

```python
kda2 = cv2.imread('kda2.jpg',0)
# plt.imshow(kda2)
img_blur = cv2.GaussianBlur(kda2, (3,3), 0)
_, binary = cv2.threshold(img_blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.imshow(binary)

kda2_str = pytesseract.image_to_string(binary)
print(kda2_str)
```
